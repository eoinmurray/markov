## Lifetimes processing

```python
from scipy.optimize import curve_fit
def func(x, a, b, c):
    return a*np.exp(-b*x) + c

f = plt.figure(figsize=(18,9))
for i in xrange(7):
    lf = np.loadtxt('A1208_08_10_13_'+repr(int(i+1))+'_lifetimes.txt', delimiter=',')
    lf[:,0] = lf[:,0]-182
    lf[:,1] = lf[:,1]/lf[:,1].max()
    
    if (i+1 is 2) or (i+1 is 4) or (i+1 is 7) or (i+1 is 1): 
        descriminator = 2.5
    else:
        descriminator = 3.2
        
    popt, pcov = curve_fit(func, lf[lf[:, 0] > descriminator,0], lf[lf[:, 0] > descriminator,1])
    
    ax = f.add_subplot(3, 3, 7-i)
    ax.set_xlim([0, 15])
    ax.set_ylim([0, 1.1])
    ax.text(0.5, 0.9, '#' + repr(int(i+1)) + ' - ' + np.array(1/popt[1]).astype('|S4').tostring() + ' ns', ha='center', va='center', transform=ax.transAxes, fontsize=16)
    ax.plot(lf[:,0], lf[:,1], 'r-')
    ax.plot(lf[:,0], func(lf[:,0], popt[0], popt[1], popt[2]), 'b--')
```
